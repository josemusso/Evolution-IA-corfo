```python
### VARIABLES NUMERICAS
# funcion para arreglar variables numericas, toma el min del intervalo (object -> int)
import re
import swifter
import numpy as np

df1 = df

def object_a_int(string):
  #print(string)
  str(string)
  string = string.replace(".", "")

  values = map(int, re.findall(r'\d+', string))
  return min(values)

# ejecutar sobre variables numéricas
start_index = df1.columns.get_loc("Ventas nacionales por tramos 2015")
end_index = df1.columns.get_loc("Empleados part time por tramos 2018")

for i in np.arange(start_index,end_index+1):
  print(df1.columns[i])
  df1.iloc[:, i] = df1.iloc[:, i].fillna("0")
  df1.iloc[:, i] = df1.iloc[:, i].astype(str).swifter.apply(object_a_int)

df1.sample(10)
```

```python
###### ENCODING A ONEHOT
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import LabelEncoder

# trabajar en copa de df_prueba para comparar despues con target encoding
df_onehot = df_prueba.copy()

# reiniciar index para los onehot
df_onehot.reset_index(drop=True, inplace=True)

### FUNCION PARA PASAR A ONEHOT Y CONCATENAR
def onehot_and_concat_to_df(df, colname):
  encoder = OneHotEncoder(categories = 'auto')
  X = encoder.fit_transform(df[colname].values.reshape(-1,1)).toarray()
  dfonehot = pd.DataFrame(X)
  dfonehot.columns = encoder.get_feature_names([colname])
  df = df.join(dfonehot)
  print(colname+': '+str(dfonehot.shape[1]) + ' columnas one-hot agregadas al dataframe')
  return df

### GRADO INNOVACION (ONEHOT)
# NA rellenado con 'No Seleccionado'
df_onehot['grado_de_innovación'] = df_onehot['grado_de_innovación'].fillna("No seleccionado")
df_onehot = onehot_and_concat_to_df(df_onehot, 'grado_de_innovación')
df_onehot.shape
df_onehot.head(10)
# eliminar columnas categoricas convertidas a onehot
del df_onehot['grado_de_innovación']

### CANALES DE VENTA (YA ESTÁ ONEHOT)
df_onehot['canales_de_venta___venta_directa_(vendedores)'].unique()
start_index = df_onehot.columns.get_loc("canales_de_venta___venta_directa_(vendedores)")
end_index = df_onehot.columns.get_loc("canales_de_venta___otro")

for i in np.arange(start_index,end_index+1):
  df_onehot.iloc[:, i] = df_onehot.iloc[:, i].map({'Seleccionado': 1, 'No seleccionado': 0})

# info redundante
del df_onehot['canales_de_venta_existentes']

### CANALES DE ADQUISISCION CLIENTES (YA ESTÁ ONEHOT)
start_index = df_onehot.columns.get_loc("canal_adquisición___a_través_de_entidades_que_reúnen_a_clientes_potenciales")
end_index = df_onehot.columns.get_loc("canal_adquisición___otros")

for i in np.arange(start_index,end_index+1):
  df_onehot.iloc[:, i] = df_onehot.iloc[:, i].map({'Seleccionado': 1, 'No seleccionado': 0})

# info redundante
del df_onehot['canal_adquisición_nuevos_clientes_y/o_usuarios']

### TIPO CLIENTES (ONEHOT)
# NA rellenado con 'No Seleccionado'
df_onehot['tipo_de_clientes'] = df_onehot['tipo_de_clientes'].fillna("No seleccionado")
df_onehot = onehot_and_concat_to_df(df_onehot, 'tipo_de_clientes')
# eliminar columnas categoricas convertidas a onehot
del df_onehot['tipo_de_clientes']


### TIPO DE PRODUCTO (ONEHOT)
# NA rellenado con 'No Seleccionado'
df_onehot['tipo_de_producto'] = df_onehot['tipo_de_producto'].fillna("No seleccionado")
df_onehot = onehot_and_concat_to_df(df_onehot, 'tipo_de_producto')
# eliminar columnas categoricas convertidas a onehot
del df_onehot['tipo_de_producto']

### SECTOR ECONOMICO (ONEHOT)
# NA rellenado con 'No Seleccionado'
df_onehot['sector_económico'] = df_onehot['sector_económico'].fillna("No seleccionado")
df_onehot = onehot_and_concat_to_df(df_onehot, 'sector_económico')
# eliminar columnas categoricas convertidas a onehot
del df_onehot['sector_económico']

### PAIS (ONEHOT)
# eliminar columnas categoricas convertidas a onehot
del df_onehot['país_procedencia_emprendimiento']

### REGION (ONEHOT)
# NA rellenado con 'No Seleccionado'
df_onehot['región'] = df_onehot['región'].fillna("No seleccionado")
df_onehot = onehot_and_concat_to_df(df_onehot, 'región')
# eliminar columnas categoricas convertidas a onehot
del df_onehot['región']

### DEMO
df_onehot.shape
df_onehot.sample(10)
```

```python
D_train = xgb.DMatrix(X_train, label=y_train)
D_test = xgb.DMatrix(X_test, label=y_test)

# modelo
# XGB Classifier
# scale to handle imbalanced dataset
scale = y_train[y_train == 0].count() / y_train[y_train == 1].count()

# xg_cl = xgb.XGBClassifier(objective='binary:logistic',
#                           nthread = 4,
#                           seed=123, 
#                           scale_pos_weight=scale,
#                           eta=0.001,
#                           max_depth=2,
#                           subsample= 0.7,
#                           colsample_bytree= 0.8,
#                           n_estimators= 300)


parameters = {
        'eta': [0.005,0.007,0.008,0.009,0.01,0.02,0.03,0.04],
        'max_depth': [5],
        'subsample': [0.6],
        'colsample_bytree': [0.6],
        'n_estimators': [60,80,100]}


xg_cl = xgb.XGBClassifier(objective='binary:logistic',
                          nthread = 4,
                          seed=123, 
                          scale_pos_weight=scale)

clf = GridSearchCV(xg_cl, parameters, n_jobs=4)


eval_set = [(X_train, y_train), (X_test, y_test)]

# Entrenar con gridsearch
clf.fit(X_train, y_train, eval_metric=["error"], eval_set=eval_set, verbose=False)
print(clf.best_params_)

```

```python
def plot_2d_space(X, y, label='Classes'):   
    colors = ['#1F77B4', '#FF7F0E']
    markers = ['o', 's']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==l, 0],
            X[y==l, 1],
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
```
